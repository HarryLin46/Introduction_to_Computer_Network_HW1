```python
import socket
import time

with open('./b08901002_p1_client_result.log', 'w') as logFile:
    logFile.write("The Client is running..\n")
    logFile.flush()

    # Configure the server IP with its corrosponding port number
    # Specify the TCP connection type and make connection to the server
    # TODO start
    HOST, PORT = '127.0.0.1', 2025
    
    #test TA's device
#     HOST, PORT = '140.112.42.104', 7777
    
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    serverName = "servername" 
    s.connect((HOST,PORT))
    print('connect complete')
    # TODO end

    Testcase = open('./p1_testcase', 'r')
    TestcaseContents = Testcase.readlines()
    Testcase.close()

    # Write the information of HOST and PORT to the client_log.txt
    # TODO start
    logFile.write("Connect to 127.0.0.1, using port number 2025\n")
    logFile.flush()
    # TODO end

    # Read test cases from p1_testcase
    # You can change the test case or create other test cases on your own
    #first we send a request
#     s.send("It's a request".encode())
    print('start for loop')
    for index,PreprossingLine in enumerate(TestcaseContents):
        Line = PreprossingLine.strip()
        print(Line)

        # For connection stability
        time.sleep(0.1)

        # Client sent the request to the server and receive the response from the server
        # TODO start
        if index % 2 == 0:
            server_respond = s.recv(1024).decode() #it should be "Please input a question for calculation"
            s.send(Line.encode())
            print("server_respond:",server_respond)
            logFile.write("Received the message from server: "+ server_respond + "\n")
            logFile.flush()

            logFile.write("Question: "+ Line + "\n")
            logFile.flush()

            answer = s.recv(1024).decode()
            print("server's answer:",answer)
            logFile.write("Answer: "+ answer + "\n")
            logFile.flush()
            
        else:
            server_respond = s.recv(1024).decode() #it should be "Do you wish to continue? (Y/N)"
            print("server_respond:",server_respond)
            logFile.write(server_respond + "\n")
            logFile.flush()
            
            s.send(Line.encode())
            logFile.write(Line + "\n")
            logFile.flush()
        
        # TODO end


    # Close the socket
    # TODO start
    s.close()
    # TODO end
logFile.close()
```
